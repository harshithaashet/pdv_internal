```python
import pandas as pd
df1 = pd.read_csv("milk_production.csv")

```

```python
df2=df1.fillna(0)
```

```python
df2.head()
```

```python
df2["y2011"] =df2["Cow Milk-2010-11"] + df2["Boffalo Milk-2010-11"] + df2["Goat Milk-2010-11"]
df2["y2012"]= df2["Cow Milk-2011-12"] + df2["Boffalo Milk-2011-12"] + df2["Goat Milk-2011-12"]
df2["y2014"]= df2["Cow Milk-2013-14"] + df2["Boffalo Milk-2013-14"] + df2["Goat Milk-2013-14"]
df2["y2015"]= df2["Cow Milk-2014-15"] + df2["Boffalo Milk-2014-15"] + df2["Goat Milk-2014-15"]
df2["y2016"]= df2["Cow Milk-2015-16"] + df2["Boffalo Milk-2015-16"] + df2["Goat Milk-2015-16"]
```

```python
print(df2.sort_values('y2014',ascending=False)["State/ UT Name"].head(1))
```

```python
print(df2.sort_values('y2011',ascending=False)["State/ UT Name"].head(5))
print(df2.sort_values('y2012',ascending=False)["State/ UT Name"].head(5))
print(df2.sort_values('y2016',ascending=False)["State/ UT Name"].head(5))
print(df2.sort_values('y2014',ascending=False)["State/ UT Name"].head(5))
print(df2.sort_values('y2015',ascending=False)["State/ UT Name"].head(5))
```
